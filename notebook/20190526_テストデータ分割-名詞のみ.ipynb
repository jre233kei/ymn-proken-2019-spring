```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
TRIAL_NUM=str(5)
```

```python
words_by_rate = [[],[],[],[],[],[]]
path = '../data/processing/raw-'+TRIAL_NUM+'.lst'
words_and_rate_raw = open(path,'r')
for wa in words_and_rate_raw:
    star = int(wa.split()[0][9])
    # print(len(wa.split()[1:]))
    words_by_rate[star].append(wa.split()[1:])
```

```python
for w in words_by_rate[5]:
    print(w)
```

```python
pd.DataFrame([w for w in words_by_rate[5]])
```

```python
dfs = []
for i in range(1,6):
    dfs.append(pd.DataFrame({
        "word" : [w for w in words_by_rate[i]],
        "star" : i
    })
    )
```

```python
df = dfs[0]
for d in dfs:
    df = df.append(d)
```

```python
df.shape
```

```python
df_train, df_test = train_test_split(df, test_size=0.3)
```

```python
for row in df_train.itertuples():

    words = row[1]
    list_element  = ""
        
    list_element  += '__label__' +str(row[2]) +','
    
    for w in words:
        if(w!="\n"):
            list_element  += ' ' + w
        
    list_element  += ' \n' 
        
    path = '../data/processing/train-'+TRIAL_NUM+'.lst'
    with open(path,mode='a') as f:
        f.write(list_element)
```
