```python
import pickle
TRIAL_NUM=9
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import MeCab

%matplotlib inline
```

```python
path = '../data/processing/train-'+str(TRIAL_NUM)+'.lst'
train =  open(path,mode='r') 

tlist = [t.split()[1:] for t in train]

words_train = []
for t in tlist:
    for tt in t:
        words_train.append(tt)
pd.DataFrame(words_train)        
```

```python
path = '../data/processing/test-'+str(TRIAL_NUM)+'.lst'
test = open(path,mode='r') 

m = MeCab.Tagger ("-Ochasen")

test_words = []
for t in test:
    m_result = m.parse(t).split('\n')[1:-2]
    for mi in m_result:
        splited_t = mi.split('\t')[0]
        test_words.append(splited_t)
```

```python
pd.DataFrame([t for t in test_words])      
```
