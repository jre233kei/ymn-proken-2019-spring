```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
words_by_rate = [[],[],[],[],[],[]]
path = '../data/processing/train.lst'
words_and_rate_raw = open(path,'r')
for wa in words_and_rate_raw:
    star = int(wa.split()[0][9])
    words_by_rate[star].extend(wa.split()[1:])
```

```python
pd.DataFrame([w for w in words_by_rate[5]])
```

```python
dfs = []
for i in range(1,6):
    dfs.append(pd.DataFrame({
        "word" : [w for w in words_by_rate[i]],
        "star" : i
    })
    )
```

```python
df = dfs[0]
for d in dfs:
    df = df.append(d)
```

```python
df_train, df_test = train_test_split(df, test_size=0.3)
```
